```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
#'id' and Unnamed: 32' column are redundant we can erase
df.info()
```

```python
#Checking total no. of null values in all features
df.isnull().sum()
```

```python
sns.countplot(x='diagnosis', data = df)
```

```python
#performing labelencoding on diagnosis column
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

```python
drop_cols = ["id", "Unnamed: 32"]

df.drop(drop_cols, axis = 1, inplace=True)
```

```python
df.head()
```

```python
df_corr = df.corr()

#features which are highly correlated with diagnosis
diagnosis_cols_corr = df_corr[df_corr['diagnosis'] > 0.7]['diagnosis']
diagnosis_cols_corr = diagnosis_cols_corr.sort_values(ascending=False)
diagnosis_cols_corr
```

```python
#We will take 4 most correlated features and see the graph
corr_cols = diagnosis_cols_corr[:5].index.values
corr_cols
```

```python
sns.pairplot(df[corr_cols],hue="diagnosis",kind="scatter")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,1:], df['diagnosis'], test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train_transformed = scaler.transform(x_train)
x_test_transformed = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

models = {1 : "Logistic Regression", 2 : "KNN classifier", 3 : "Decision Tree Classifier" , 4 : "Random Forest Classifier"}
models_score = {}

model1 = LogisticRegression(random_state=0)
model1.fit(x_train,y_train)
models_score[1] = model1.score(x_test,y_test) * 100

model2 = KNeighborsClassifier(n_neighbors = 1)
model2.fit(x_train,y_train)
models_score[2] = model2.score(x_test,y_test) * 100

model3 = DecisionTreeClassifier(random_state=0)
model3.fit(x_train,y_train)
models_score[3] = model3.score(x_test,y_test) * 100

model4 = RandomForestClassifier(n_estimators = 100)
model4.fit(x_train,y_train)
models_score[4] = model4.score(x_test,y_test) * 100
```

```python
max_score, ind  = -1 , -1
for i, score in models_score.items():
    if score > max_score:
        max_score = score
        ind = i
print("Best model is {} with accuracy score {}".format(models[ind],max_score))
```

# Random Forest Classifier is the best model among all four

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model4.predict(x_test)

print('Accuracy is: ', accuracy_score(y_test,y_pred))
print('Confusion Matrix is: ', confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```
